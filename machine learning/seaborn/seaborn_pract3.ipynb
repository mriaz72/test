```python
import seaborn as sns
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.barplot()
```

```python
sns.barplot(x=tips_df.day,y=tips_df.tip)
```

```python
sns.barplot(x=tips_df.day,y=tips_df.tip,hue=tips_df.sex)
```

```python
sns.barplot(data=tips_df,x="day",y="tip",hue="sex")
```

```python
sns.barplot(data=tips_df,x="day",y="tip",hue="sex",palette="hot")
```
