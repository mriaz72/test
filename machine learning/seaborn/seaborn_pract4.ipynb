```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
t_df = sns.load_dataset("titanic")
t_df
```

```python
sns.scatterplot(x="age",y="fare",data=t_df)
```

```python
sns.scatterplot(x="age",y="fare",data=t_df,hue="sex")
```

```python
sns.scatterplot(x="age",y="fare",data=t_df,hue="sex",style="who")
```

```python
sns.scatterplot(x="age",y="fare",data=t_df,hue="sex",style="who",size="who")
```

```python
sns.scatterplot(x="age",y="fare",data=t_df,hue="sex",style="who",size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=t_df,hue="sex",style="who",size="who",sizes=(100,400))
```

```python
#to see women are more alive or mem or children
plt.figure(figsize=(16,9))
sns.scatterplot(x="who",y="fare",data=t_df,hue="alive",style="alive",size="who",sizes=(100,400))
```
